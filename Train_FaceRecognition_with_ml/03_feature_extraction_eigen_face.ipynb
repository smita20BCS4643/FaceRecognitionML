### Feature Extraction with Eigen Face



```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt

import cv2

#Eigen Face
from sklearn.decomposition import PCA
import pickle
```

```python
#load the data 
data = pickle.load(open('./data/data_images_100_100.pickle',mode='rb')) #load the data
```

```python
data.head()
```

### Eigen Face

```python
#mean face
X = data.drop('gender',axis =1).values #all images
```

```python
X
```

```python
mean_face = X.mean(axis=0) # falttern the pattern
mean_face.shape
```

```python
#visualization  of mean faces
plt.imshow(mean_face.reshape((100,100)),cmap = 'gray')
plt.axis('off')
plt.show()
```

### substract with mean face


Apply X_t data to PCA

@ find the right number of componants- elbow
@ with right number compnent computer principal components

```python
X_t = X - mean_face #transformend data
```

```python
pca = PCA(n_components=None, whiten = True,svd_solver= 'auto')
pca.fit(X_t)
```

```python
exp_var_df = pd.DataFrame()
exp_var_df['explained_var'] = pca.explained_variance_ratio_
exp_var_df['cum_explained_var']=exp_var_df['explained_var'].cumsum()
exp_var_df['principle_components']=np.arange(1,len(exp_var_df)+1)

exp_var_df.head()
```

```python
exp_var_df.set_index('principle_components',inplace=True)
```

```python
#visualiza explained variance
fig, ax = plt.subplots(nrows=2,figsize=(15,12))

exp_var_df['explained_var'].head(100).plot(kind ='line',marker='o',ax = ax[0])
exp_var_df['cum_explained_var'].head(100).plot(kind='line',marker='o',ax= ax[1])
```

from above plot , it is clear that with approximately 50 principal components having 80% converage

```python
pca_50 = PCA(n_components=50, whiten=True,svd_solver= 'auto')
pca_data = pca_50.fit_transform(X_t)
```

```python
pca_data.shape
```

```python
#saving  data and the model
y = data['gender'].values # independent variable
np.savez('./data/data_pca_50_target',pca_data,y)
pca_dict = {'pca':pca_50,'mean_face':mean_face}
```

```python
pickle.dump(pca_dict,open('./model/pca.dict.pickle','wb'))
```

Visualize Eigen Image

```python
pca_data_inv = pca_50.inverse_transform(pca_data)
```

```python
pca_data_inv.shape
```

```python
eig_img = pca_data_inv[0,:].reshape((100,100))
eig_img.shape
```

```python
plt.imshow(eig_img,cmap='gray')
plt.axis('off')
```
